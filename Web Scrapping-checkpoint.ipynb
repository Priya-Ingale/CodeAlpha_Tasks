```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=sunglass&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
brand=[]
description =[]
price=[]
```

```python
names = soup.find_all('div',class_="Fo1I0b")
for i in names:
    name=i.text
    brand.append(name)
brand=brand[0:40]
print(brand)
len(brand)
```

```python
desc = soup.find_all('a',class_="atJtCj")

for i in desc:
    line=i.text
    description.append(line)
description = description[0:40]
print(description)
len(description)
```

```python
Price = soup.find_all('div',class_="hZ3P6w")

for i in Price:
    number=i.text
    price.append(number)
price=price[0:40]
print(price)
len(price)
```

```python
df=pd.DataFrame({"Brand Name": brand, "Product Description": description, "Price": price})
print(df)
```
